Faire une recherche automatique sur plusieurs journaux

```python
# Définir une liste de liens
journaux = ["https://www.letemps.ch/","https://24heures.ch/"]

# Boucle sur les liens
for journal in journaux:
    
    #Obtenir la page
    page = requests.get(journal)
    
    # Afficher le nombre de fois où le mot France est présent
    print(page.text.count("France"))
```
